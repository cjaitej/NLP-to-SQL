```python
import random
import json

with open("bird_dev_all.json", "r") as f:
    unique = json.load(f)

# pick 200 random (without repetition)
first_200 = random.sample(unique, 200)

# the rest are everything not in first_200
first_200_set = {json.dumps(x, sort_keys=True) for x in first_200}
rest = [x for x in unique if json.dumps(x, sort_keys=True) not in first_200_set]

print(len(first_200), len(rest))

with open("bird_200.json", "w") as f:
    json.dump(first_200, f, indent=2)

with open("bird_rest.json", "w") as f:
    json.dump(rest, f, indent=2)
```
